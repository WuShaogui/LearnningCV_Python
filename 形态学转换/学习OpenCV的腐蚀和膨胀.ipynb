```python
#查看图片
import matplotlib.pyplot as plt
def show_images(images,closize=2,set_locator=False,titles=[]):
    # 计算行数量
    rows=int(len(images)/closize)+1
    
    fig=plt.figure(figsize=(34,14))
    for find,image in enumerate(images):
        ax=fig.add_subplot(rows,closize,find+1) #每个子图
        ax.imshow(image)
        
        # 标题
        if len(titles)==0:
            plt.title('{}-({})'.format(find,image.shape)) #图片大小作为标题
        else:
            plt.title('{}-({})'.format(titles[find],image.shape)) #自定义标题
        
        # 标度
        h,w=image.shape[:2]
        plt.yticks(range(h),[str(i) for i in range(1,h+1)])
        plt.xticks(range(w),[str(i) for i in range(1,w+1)])
        
        # 加像素值
        for row_ind in range(len(image)):
            for clo_ind in range(len(image[row_ind])):
                plt.text(row_ind,clo_ind,image[row_ind][clo_ind])
    plt.show()
```

```python
import cv2,sys
import numpy as np
img = np.zeros((10, 10),np.uint8)
for row in range(10):
    for col in range(10):
        img[row,col] = row*col

kernalInt = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
imgErode = cv2.erode(img, kernalInt, borderType=cv2.BORDER_CONSTANT, borderValue=0)
imgDilate = cv2.dilate(img, kernalInt, borderType=cv2.BORDER_CONSTANT, borderValue=0)
show_images([img,kernalInt,imgErode,imgDilate],titles=['original','kernel','imgErode','imgDilate'],closize=4)
```
